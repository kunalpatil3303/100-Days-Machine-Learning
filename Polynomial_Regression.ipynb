```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
```

```python
X = 6 * np.random.rand(200, 1) - 3
Y = 0.8 * X**2 + 0.9 * X + 2 + np.random.randn(200, 1)

# y = 0.8x^2 + 0.9x + 2
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
ploy = PolynomialFeatures(degree=2)
X_train = ploy.fit_transform(X_train)
X_test = ploy.transform(X_test)
```

```python
model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
X_test_prediction = model.predict(X_test)
X_test_r2_score = r2_score(X_test_prediction, Y_test)
print(X_test_r2_score)
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(X,Y)
plt.show()
```

```python
idx = np.argsort(X_train[:, 1])
X_sorted = X_train[idx]

y_sorted = model.predict(X_sorted)

plt.figure(figsize=(10, 5))
plt.scatter(X, Y)
plt.plot(X_sorted[:, 1], y_sorted, color='red', linewidth=2)
plt.show()
```
