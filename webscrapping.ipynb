```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}

response = requests.get("https://webscraper.io/test-sites/e-commerce/allinone", headers=headers).text
```

```python
soup = BeautifulSoup(response, 'html.parser')
```

```python
print(soup.prettify())
```

```python
item = soup.find_all("div", class_="product-wrapper card-body")
```

```python
len(item)
```

```python
name = []
price = []
description = []
for i in item:

    name.append(i.find("a", class_="title").text.strip())
    price.append(i.find("h4", class_="price float-end card-title pull-right").text.strip())
    description.append(i.find('p', class_="description card-text").text.strip())
```

```python
df = pd.DataFrame(
    {
        'Name': name,
        'Price': price,
        'Description': description
    }
)
```
