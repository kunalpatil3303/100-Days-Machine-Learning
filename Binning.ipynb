```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import KBinsDiscretizer
from sklearn.model_selection import cross_val_score
```

```python
df = pd.read_csv("/content/Titanic-Dataset.csv",usecols=["Age","Fare","Survived"])
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
X = df.drop(columns = ["Survived"])
Y = df["Survived"]
```

```python
print(X)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train, Y_train)
```

```python
X_test_prediction = model.predict(X_test)
X_test_accuracy = accuracy_score(X_test_prediction, Y_test)
print(X_test_accuracy)
```

```python
np.mean(cross_val_score(DecisionTreeClassifier(),X,Y,cv=10, scoring="accuracy"))
```

```python
trf = ColumnTransformer([
    ("kbin_age", KBinsDiscretizer(n_bins = 10, encode = "ordinal", strategy= "quantile"),[0]),
    ("kbin_fare", KBinsDiscretizer(n_bins = 10, encode = "ordinal", strategy = "kmeans"), [1])
], remainder = "passthrough")
```

```python
X_train_trf = trf.fit_transform(X_train)
X_test_trf = trf.transform(X_test)
```

```python
model.fit(X_train_trf, Y_train)
```

```python

X_test_prediction = model.predict(X_test_trf)
X_test_accuracy = accuracy_score(X_test_prediction, Y_test)
print(X_test_accuracy)
```

```python
np.mean(cross_val_score(DecisionTreeClassifier(),X_train_trf,Y_train,cv=10, scoring="accuracy"))
```
