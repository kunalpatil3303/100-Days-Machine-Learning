```python
upper_limit = df["Height"].quantile(0.99)
lower_limit = df["Height"].quantile(0.01)
```

```python
new_df = df[(df["Height"] < upper_limit) & (df["Height"] > lower_limit)]
```

```python
new_df.shape
```

```python
df["Height"] = np.where(df["Height"] > upper_limit, upper_limit, df["Height"])
df["Height"] = np.where(df["Height"] < lower_limit, lower_limit, df["Height"])
```
